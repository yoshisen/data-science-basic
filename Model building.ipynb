```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import make_regression, make_blobs
from sklearn.preprocessing import StandardScaler
from matplotlib import rcParams

rcParams['font.family'] = 'Meiryo'
```

```python
# 線形回帰（Linear Regression）
# データの生成
X_reg, y_reg = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

# モデルの作成と学習
model_reg = LinearRegression()
model_reg.fit(X_reg, y_reg)

# 予測
y_pred = model_reg.predict(X_reg)

# 可視化
plt.figure(figsize=(6, 4))
plt.scatter(X_reg, y_reg, label="実データ")
plt.plot(X_reg, y_pred, color='red', label="予測")
plt.title("線形回帰の結果")
plt.xlabel("特徴量")
plt.ylabel("ターゲット")
plt.legend()
plt.show()
```

```python
# K-means クラスタリング
# データの生成
X_clust, _ = make_blobs(n_samples=200, centers=3, random_state=42)

# モデルの作成と学習
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X_clust)
labels = kmeans.labels_

# 可視化
plt.figure(figsize=(6, 4))
plt.scatter(X_clust[:, 0], X_clust[:, 1], c=labels, cmap='viridis', marker='o', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label="クラスタ中心")
plt.title("K-means クラスタリング")
plt.xlabel("特徴量1")
plt.ylabel("特徴量2")
plt.legend()
plt.show()
```

```python
#  主成分分析（PCA）
# データの生成
X_pca = np.random.rand(100, 5) * 10  # 5次元データ
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_pca)

# PCAの適用
pca = PCA(n_components=2)
X_pca_2d = pca.fit_transform(X_scaled)

# 可視化
plt.figure(figsize=(6, 4))
plt.scatter(X_pca_2d[:, 0], X_pca_2d[:, 1], alpha=0.7)
plt.title("PCA による次元削減")
plt.xlabel("主成分1")
plt.ylabel("主成分2")
plt.show()
```
