```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

rcParams['font.family'] = 'Meiryo'

# データの生成（例としてランダムなデータを作成）
np.random.seed(42)
data = pd.DataFrame({
    '年齢': np.random.randint(20, 60, 100),
    '年収': np.random.randint(300, 1000, 100) * 1000,
    '購入回数': np.random.randint(1, 20, 100)
})

# データの基本情報の表示
print("データの概要:")
print(data.describe())
print("\nデータの前5行:")
print(data.head())

# ヒストグラム（データの分布）
plt.figure(figsize=(12, 4))
sns.histplot(data['年収'], bins=20, kde=True)
plt.title("年収の分布")
plt.xlabel("年収（万円）")
plt.ylabel("頻度")
plt.show()

# 散布図（相関関係の可視化）
plt.figure(figsize=(6, 4))
sns.scatterplot(x=data['年齢'], y=data['年収'])
plt.title("年齢と年収の関係")
plt.xlabel("年齢")
plt.ylabel("年収（万円）")
plt.show()

# 箱ひげ図（外れ値の検出）
plt.figure(figsize=(6, 4))
sns.boxplot(x=data['購入回数'])
plt.title("購入回数の分布（箱ひげ図）")
plt.xlabel("購入回数")
plt.show()

# 相関係数のヒートマップ
data_corr = data.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(data_corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("相関係数ヒートマップ")
plt.show()
```
